```python
def calculate_order_total(price, tax_rate, quantity):
    """Tính tổng giá trị đơn hàng bao gồm thuế.
    
    Args:
        price (float): Giá của một sản phẩm
        tax_rate (float): Tỷ lệ thuế (ví dụ: 0.1 cho 10%)
        quantity (int): Số lượng sản phẩm
    
    Returns:
        float: Tổng giá trị đơn hàng sau thuế
    """
    subtotal = price * quantity
    return subtotal * (1 + tax_rate)





```

```python
def process_user_data(user_list):
    """Xử lý dữ liệu người dùng và lọc theo domain email.
    
    Args:
        user_list (list): Danh sách các dictionary chứa 'name' và 'email'
    
    Returns:
        list: Danh sách tên người dùng viết hoa có email thuộc domain 'example.com'
    """
    filtered_names = []
    for user in user_list:
        if user['email'].endswith('@example.com'):
            filtered_names.append(user['name'].upper())
    return filtered_names

# Tạo dữ liệu mẫu để test function
sample_users = [
    {'name': 'Nguyễn Văn An', 'email': 'an@example.com'},
    {'name': 'Trần Thị Bình', 'email': 'binh@gmail.com'},
    {'name': 'Lê Hoàng Cường', 'email': 'cuong@example.com'},
    {'name': 'Phạm Thị Dung', 'email': 'dung@yahoo.com'},
    {'name': 'Võ Minh Tuấn', 'email': 'tuan@example.com'}
]

# Test function với dữ liệu mẫu
result = process_user_data(sample_users)
print("Kết quả lọc users có email @example.com:")
for name in result:
    print(f"- {name}")

print(f"\nTổng cộng: {len(result)} users được lọc")

```

```python
def calculate_discounted_price(price, discount_rate):
    """
    Tính giá sau khi giảm giá.
    
    PHÂN TÍCH LỖI LOGIC:
    - Hàm này có tên là "calculate_discounted_price" (tính giá sau giảm)
    - Nhưng lại return "discount_amount" (số tiền được giảm)
    - Đây là lỗi logic: trả về sai giá trị
    
    VÍ DỤ LỖI:
    - Giá gốc: 100,000 VND
    - Discount rate: 0.1 (10%)
    - Hàm hiện tại trả về: 10,000 VND (số tiền giảm)
    - Nhưng tên hàm gợi ý phải trả về: 90,000 VND (giá sau giảm)
    
    CÁCH SỬA:
    - Cách 1: Đổi tên hàm thành "calculate_discount_amount"
    - Cách 2: Sửa logic để trả về giá sau giảm (khuyến nghị)
    """
    discount_amount = price * discount_rate
    # SỬA LỖI: Trả về giá sau khi trừ đi số tiền giảm
    discounted_price = price - discount_amount
    return discounted_price

# Demo để thấy sự khác biệt:
print("=== DEMO PHÂN TÍCH LỖI ===")
price = 100000
discount_rate = 0.1

# Hàm đã sửa (đúng)
final_price = calculate_discounted_price(price, discount_rate)
print(f"Giá gốc: {price:,} VND")
print(f"Tỷ lệ giảm: {discount_rate*100}%")
print(f"Giá sau giảm: {final_price:,} VND")

# Nếu dùng logic cũ (sai)
old_logic_result = price * discount_rate
print(f"Logic cũ sai trả về: {old_logic_result:,} VND (chỉ là số tiền giảm)")
```

```python
def get_active_users(users):
    """
    Lọc danh sách users để chỉ lấy những user đang hoạt động.
    
    PHÂN TÍCH TỐI ƯU HÓA:
    - Code cũ: Dùng range(len()) và indexing - không Pythonic
    - Code cũ: So sánh == True thừa thải
    - Code cũ: Dùng loop và append thay vì list comprehension
    
    CÁCH TỐI ƯU:
    - Dùng iteration trực tiếp thay vì indexing
    - Bỏ == True vì is_active đã là boolean
    - Dùng list comprehension cho code ngắn gọn hơn
    """
    return [user for user in users if user.is_active]
```

```python
def divide(a, b):
    """
    Thực hiện phép chia với xử lý lỗi.
    
    PHÂN TÍCH LỖI:
    - Chia cho 0 sẽ gây ra ZeroDivisionError
    - Không có xử lý ngoại lệ
    
    CÁCH SỬA:
    - Thêm try-except để bắt ZeroDivisionError
    - Trả về thông báo lỗi hoặc giá trị mặc định
    """
    try:
        return a / b
    except ZeroDivisionError:
        print(f"Lỗi: Không thể chia {a} cho 0")
        return None

result = divide(10, 0)
if result is not None:
    print(f"Kết quả: {result}")
else:
    print("Phép chia không thực hiện được")
```
